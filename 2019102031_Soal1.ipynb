```python
#NO 2
import pandas as pd
import numpy as np
from functools import reduce
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset_uts-2021.csv')
df.head()
```

```python
#a. Menghapus Baris yang berulang
df.drop_duplicates(inplace = True)
print(df.to_string()) 
```

```python
#b. Menghapus Baris yang hilang (Ada Nilai NaN nya)
df.dropna(inplace = True)

print(df.to_string())
```

```python
#c. Menghapus kolom yang berisi data yang tidak releva (Hapus Kolom No)
to_drop = ['No']
df.drop(to_drop, inplace = True, axis = 1)
print(df.to_string())
```

```python
#d. Perbaiki kesalahan input (intenal harusnya Internal) 
df.loc[4, 'Project'] = 'Internal'
print(df.to_string())
```

```python
#e. Ubah Tipe data float64 untuk Mobile Phone menjadi Teks atau Object
df = df.astype({"Mobile Phone": str})
df.dtypes
```

```python
#f. Ganti nama semua kolom dalam bahasa Indonesia 
df.rename(columns={'Name':'Nama','Address':'Alamat','Mobile Phone':'Nomor Telepon','Project':'Projek'}, inplace=True)
print(df.to_string())
```
